## Ejercicio 1: Pirámide

a) En este ejercicio de deberá construir una pirámide negra sobre un fondo blanco, debe tener el siguiente aspecto:

<img src="piramide.png">

**Pista 1:** Dicha pirámide puede ser construida de la misma manera que construimos la cruz en clase. Utilice cuadrados negros y blancos de tamaño de 64 x 64.

**Pista 2:** El tamaño que debe poseer la pirámide es de 192 x 320.
b) muestre la imagen generada utilizando la librería `matplotlib` (recuerde que debe importar la librería previamente)

```python
import numpy as np
import matplotlib.pyplot as plt

negro=np.zeros((64,64))
blanco=1-negro

Primerafila=np.concatenate((blanco,blanco,negro,blanco,blanco), axis=1)
Segundafila=np.concatenate((blanco,negro,negro,negro,blanco), axis=1)
Tercerafila=np.concatenate((negro,negro,negro,negro,negro), axis=1)

piramide=np.concatenate((Primerafila, Segundafila, Tercerafila), axis=0)


plt.imshow(piramide, cmap='gray')
plt.show()
```

## Ejercicio 2: Carga y Edición de Imagen

a) Cargue una imagen a color utilizando la librería `PIL` y muestre la misma utilizando `matplotlib` (recuerde que debe importar las librerías previamente).

c) Transformar la imagen de color a escala de grises y mostrar la misma utilizando `matplotlib`.

```python
from PIL import Image
from matplotlib.pyplot import imshow
import numpy as np

#Carga de la imagen con PIL
G=Image.open('globos.jpg')

#muestra de la imagen con matplotlib
G_array3D=np.array(G)
fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True)
fig.set_figheight(8)
fig.set_figwidth(10)
axs[0].imshow(G_array3D)
axs[0].set_title('Imagen Globos a color')
#imshow(G_array3D)

#Tamaño, modo y formato de la imagen.
print("Tamaño=", G.size,"formato=", G.format," y tipo=", G.mode)
G.show()

#Imagen en escala de grises
G_gris=G.convert('L')
G_array3D_gris=np.array(G_gris)
axs[1].set_title('Imagen Globos en escala de grises')
axs[1].imshow(G_array3D_gris, cmap='gray')


```

## Ejercicio 3: Formatos de Imagen

a) Cargue una imagen a color utilizando la librería `PIL` (recuerde que debe importarla previamente).

b) Obtenga tres imágenes arreglos utilizando la librería `numpy` : las componentes colorada, verde y azul (RGB) de la misma imagen. 

c) Grafique los tres arreglos o componentes por separado utilizando la librería `matplotlib` (recuerde que debe importarla previamente). Imprima el tamaño de cada una de las componentes utilizando el método `shape`.

d) ¿Qué puede decir al respecto del tamaño de cada una de las componentes? Explique por qué ocurre dicha coincidencia. 

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

#a) Carga y muestra de imagen
C=Image.open('coco.jpg')
C_array3D=np.array(C)
imshow(C_array3D)
plt.title('Imagen COCO con libreria PIL')
plt.show()
print()

#b) arreglos de componentes 
#componente roja
C_array3D_rojo=C_array3D[:,:,0]
#componente verde
C_array3D_verde=C_array3D[:,:,1]
#componente azul
C_array3D_azul=C_array3D[:,:,2]

#Grafico de los arreglos de las componentes

#ROJA
plt.imshow(C_array3D_rojo, cmap='gray')
plt.title('Imagen COCO, componente ROJA')
plt.colorbar()
plt.show()
print("Tamaño de la componente roja", C_array3D_rojo.shape)

#VERDE
plt.imshow(C_array3D_verde, cmap='gray')
plt.title('Imagen COCO, componente VERDE')
plt.colorbar()
plt.show()
print("Tamaño de la componente verde", C_array3D_verde.shape)

#AZUL
plt.imshow(C_array3D_azul, cmap='gray')
plt.title('Imagen COCO, componente AZUL')
plt.colorbar()
plt.show()
print("Tamaño de la componente azul", C_array3D_azul.shape)

```

d) ¿Qué puede decir al respecto del tamaño de cada una de las componentes? Explique por qué ocurre dicha coincidencia.

Dicha coincidencia ocurre debido a que una imagen RGB, se almacena como un arreglo de datos de (MxNx3), que define los componentes de color rojo, verde y azul para cada pixel individual. Por lo cual el tamaño de la matriz para las 3 capas es la misma, lo que varia es la intensidad del pixel en cada capa, que va a formar el color cuando se combinen las 3.
